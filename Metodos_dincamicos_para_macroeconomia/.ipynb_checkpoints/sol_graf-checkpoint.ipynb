```python
import math
import numpy as np
from scipy.special import eval_hermite as hn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['figure.figsize']=10,6
```

En la primera parte veremos gráficamente cómo se comportan las soluciones de las ecuaciones en diferencias encontradas en las sesiones de estudio.

### Ejemplo 1

$$
x_{t} = -\left( -\frac{1}{2} \right)^{t} + 2 
$$

```python
def ed_1(t):    
    S = -np.power(-0.5,t) + 2    
    return S
```

```python
a = 2
periods = 10
x_coordinates = [0, periods]
y_line = [a, a]
t = np.linspace(start=0, stop=periods, num=periods+1)
```

```python
plt.plot(x_coordinates, y_line , color = 'blue', linewidth = 1,
         label = r'$b=2$')
plt.plot(t, ed_1(t) , color = 'gray', linewidth = 2,
         linestyle = (0, (5, 10)))
plt.scatter(t, ed_1(t) , color = 'red', linewidth = 3 , 
            marker = 'o' , s = 25 ,
            label = r'$x_{t}$')
plt.title(r'$x_{t} = -\left( -\frac{1}{2} \right)^{t} + 2 $', 
          fontsize = 18 , color = 'black')
plt.ylabel(r'$x_t$' , fontsize = 17 , 
           color = 'black')
plt.xlabel(r'$t$' , fontsize = 17, 
           color = 'black')
plt.yticks(color = 'black', fontsize = 14)
plt.xticks(color = 'black', fontsize = 14)
plt.grid(color = 'pink' , alpha = 0.8)
plt.legend(loc='best' , facecolor = 'ivory' , 
           edgecolor = 'black' , 
           fontsize = 13)
plt.show()
```

### Ejemplo 2:

$$
x_{t} = -3\left( \frac{2}{3} \right)^{t} + 4 
$$

```python
a = 4
periods = 10
x_coordinates = [0, periods]
y_line = [a, a]
t = np.linspace(start=0, stop=periods, num=periods+1)
```

```python
def ed_2(t):    
    S = -3*np.power(2/3,t) + 4  
    return S
```

```python
plt.plot(x_coordinates, y_line , color = 'blue', linewidth = 1,
         label = r'$b=2$')
plt.plot(t, ed_2(t) , color = 'gray', linewidth = 2,
         linestyle = (0, (5, 10)))
plt.scatter(t, ed_2(t) , color = 'red', linewidth = 3 , 
            marker = 'o' , s = 25 ,
            label = r'$x_{t}$')
plt.title(r'$x_{t} = -3\left( \frac{2}{3} \right)^{t} + 4 $', 
          fontsize = 18 , color = 'black')
plt.ylabel(r'$x_t$' , fontsize = 17 , 
           color = 'black')
plt.xlabel(r'$t$' , fontsize = 17, 
           color = 'black')
plt.yticks(color = 'black', fontsize = 14)
plt.xticks(color = 'black', fontsize = 14)
plt.grid(color = 'pink' , alpha = 0.8)
plt.legend(loc='best' , facecolor = 'ivory' , 
           edgecolor = 'black' , 
           fontsize = 13)
plt.show()
```
